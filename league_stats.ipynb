```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_html('standings.html')
```

```python
stats = df1[4]
```

```python
stats.drop([0,1,2],inplace=True)
```

```python
columns = ['team','R','TB','RBI','BB-H','K','HBP','SB','CS','IP','H','ER','BB-P','HB','K-P','QS','W','L','SV','PF','PA','STRK','Moves']
```

```python
stats.drop([0,2,22,11],axis=1,inplace=True)
```

```python
stats.columns=columns
```

```python
for c in stats.columns:
  if stats[c].dtype == 'object':
    try:
        stats[c] = pd.to_numeric(stats[c])
    except:
        pass
```

```python
wp = {'Mr. May':.789,'The Cat Barbers':.632,'Lucky Gravy':.526,'HIGH HEAT':.316,'Dong Juan':.316,'The Grundys':.579,'Fungo Kings':.526,'Michigan and Trumbull':.526,'Los Ninos de Rigoberto':.421,'GR Mudrats':.368}
```

```python
stats['winning %'] = stats['team'].map(wp)
```

```python
#stats.set_index('team')
```

```python
corr = stats.corr()
```

```python
f, ax = plt.subplots(figsize=(15, 15))
```

```python
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,
            ax=ax,
            annot=False,
            cmap='cool')
```
